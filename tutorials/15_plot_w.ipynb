```python
import numpy as np
import matplotlib.pyplot as plt
import cycler

ndmft   = 10
arr_idmft = range(1,1+ndmft,2)

fig, axarr = plt.subplots( 2,2 )
axarrflat   = axarr.flatten()
axi = axarr[0,0]
axr = axarr[0,1]
axi.set_prop_cycle( cycler.cycler('color', [plt.cm.coolwarm(i) for i in np.linspace(0, 1, ndmft)]) )
axr.set_prop_cycle( cycler.cycler('color', [plt.cm.coolwarm(i) for i in np.linspace(0, 1, ndmft)]) )

dirnameData = "./data/"

for idmft in arr_idmft :
    dat = np.genfromtxt(dirnameData+"Gw_1_1_i{}.dat".format(idmft) ).transpose()
    # print dat[0]
    axr.plot( dat[0], dat[1], "-", label="i={}".format(idmft), alpha=1.0, markersize=idmft )
    axi.plot( dat[0], dat[2], "-", label="i={}".format(idmft), alpha=1.0, markersize=idmft )



ax2i = axarr[1,0]
ax2r = axarr[1,1]
ax2i.set_prop_cycle( cycler.cycler('color', [plt.cm.coolwarm(i) for i in np.linspace(0, 1, ndmft)]) )
ax2r.set_prop_cycle( cycler.cycler('color', [plt.cm.coolwarm(i) for i in np.linspace(0, 1, ndmft)]) )

for idmft in range(1,1+ndmft) :
    dat = np.genfromtxt(dirnameData+"Selfw_1_1_i{}.dat".format(idmft) ).transpose()
    # print dat[0]
    ax2r.plot( dat[0], dat[1], "-", label="R,i={}".format(idmft), alpha=1.0, markersize=idmft )
    ax2i.plot( dat[0], dat[2], "-", label="I,i={}".format(idmft), alpha=1.0, markersize=idmft )



# plt.xlim( -2,2 )
# for axnow in axarrflat :
#     axnow.set_xlim( 0, 2 )
axi.legend()
plt.show()
```
